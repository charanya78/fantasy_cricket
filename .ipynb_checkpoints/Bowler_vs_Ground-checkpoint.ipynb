```python
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np 
import os
import pandas as pd
```

```python
df1 = pd.read_csv('archive/deliveries.csv')
df1.dataframeName = 'deliveries.csv'
```

```python
df1.head(5)
```

```python
len(df1)
```

```python
player_subset = df1[df1["bowler"] == 'JJ Bumrah']
player_subset
```

```python
df2 = pd.read_csv('archive/matches.csv')
df2.dataframeName = 'matches.csv'
```

```python
df2.head(5)
```

```python
venue_subset = df2[df2["city"] == 'Bangalore']
venue_subset
```

```python
match_id_list = venue_subset["id"].to_numpy()
print(match_id_list)
```

```python
m = len(match_id_list)
```

```python
player_subset
```

```python
final_df = []
for i in range(len(player_subset)):
    if player_subset.iat[i,0] in match_id_list:
        final_df.append(player_subset.iloc[i])
```

```python
final_df
```

```python
final_df = pd.DataFrame(final_df)
```

```python
len(final_df)
```

```python
#Overall

total_innings = len(pd.unique(final_df['match_id']))
print(total_innings)

total_runs = final_df["batsman_runs"].sum() + final_df["wide_runs"].sum() + final_df["noball_runs"].sum()
print(total_runs)

total_outs = final_df["player_dismissed"].isnull().sum()
total_outs = len(final_df)-total_outs
print(total_outs)

average = total_runs/total_outs
print(average)

strike_rate = len(final_df)/ total_outs
print(strike_rate)

economy = total_runs / (len(final_df)/6)
print(economy)
```
