```python
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np 
import os
import pandas as pd
```

```python
df = pd.read_csv('archive/matches.csv')
df.dataframeName = 'matches.csv'
```

```python
season_subset = df[(df["season"] == 2017 )]
season_subset
```

```python
match_id_list = season_subset["id"].to_numpy()
print(match_id_list)
```

```python
df1 = pd.read_csv('archive/deliveries.csv')
df1.dataframeName = 'deliveries.csv'
```

```python
team_subset = df1[df1["bowling_team"] == 'Delhi Daredevils']
team_subset
```

```python
final_df = []
for i in range(len(team_subset)):
    if team_subset.iat[i,0] in match_id_list:
        final_df.append(team_subset.iloc[i])
```

```python
final_df = pd.DataFrame(final_df)
final_df
```
